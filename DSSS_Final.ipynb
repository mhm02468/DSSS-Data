**Literature Review**

Extensive interdisciplinary literature draws correlation between social outcomes, education policy and academic attainment in American education institutions. In fact, an increasing proportion of related studies mobilise Machine Learning methods in tracing relationships between these factors. Many discuss Educational Data Mining (EDM), an application of data mining techniques to analyse educational data (Ahmed and Elaraby, 2014; Burgos et al., 2018; Witten and Frank, 2002). In one article, Shete et al (2022) propose a unique machine learning algorithm to estimate student outcomes in degree programs, addressing challenges in student histories, course choices, and growth predictions. The technique uses a bi-layered design and data-driven strategy, outperforming benchmark methods in simulations with undergraduate student data from UCLA (Shete et al., 2022, 1949). They define academic success in terms of students' ability to graduate on time and achieve satisfactory Grade Point Averages (GPAs). While most literature focuses on higher education, a similar study applies these methods to model ‘success’ at the elementary through high school level (K-12). Harvey and Kumar (2019) use standardised testing data and three classifiers - linear regression, decision tree, and Naive Bayes techniques - for developing predictive models (Harvey and Kumar, 2019, 3004). Their study showed strong correlations between socioeconomic factors and SAT scores in Massachusetts. Further, they concluded that average expenditure per pupil was negatively correlated with average SAT math scores (Harvey and Kumar, 2019, 3009). Further, of their models, the decision tree and Naive Bayes models were the most effective in predicting SAT math test scores, with Naive Bayes having an accuracy of 71.0% (Harvey and Kumar, 2019, 3010). 

**Data**

Although test scores, graduation rates, and GPAs are useful measures of attainment, they may not fully capture a student's potential or predict future success on their own. And, in fact, the 2017 Massachusetts Public Schools data that is used in this report includes extensive information on these categoties. However, this study takes two alternative metrics published by the Massachusetts Department of Elementary and Secondary Education: Accountability Percentile and the Cumulative Progress and Performance Index (PPI). The **Accountability Percentile** measures a school's overall performance in comparison to other schools. The percentile of a school accounts for various factors such as achievement, growth, progress of English learners, chronic absenteeism, completion rates of high school and advanced coursework. Schools that score higher percentiles are generally considered to be high-performing, whereas those with lower percentiles are considered to be low-performing (CITE).The Cumulative **Progress and Performance Index (PPI)** is a number between 0 and 100 that combines information about how well a school or district is doing in areas like student growth, graduation rates, and narrowing proficiency gaps. It is calculated over a four-year period, using annual PPIs based on data from the previous two years. The most recent years are weighted the most heavily. The cumulative PPI is always reported on a 100-point scale, and a cumulative PPI of 75 or higher for both "all students" and high-needs students indicates that a school is making progress toward narrowing proficiency gaps (CITE). Thus, these two measures provide a unique window into how we understand success; they are more holistic and provide unique insights. The former, as a comparative metric, provides a conceptual standardisation of how we understand success. The latter is more focused insofar as it encasuplates a wider array of criteria. # Methods

## <u>General Approach<u>

Importantly, it does the preprocessing, PCA, and regressing in a pipeline; this reduces the likelihood of data leakage and streamlines the process.

## <u> Supervised: Regressors <u>

Dimensionality reduction techniques are often unsupervised learning methods employed to reduce the number of features in a dataset, particularly when the number of features is large relative to the number of observations. Linear dimensionality reduction techniques, such as principal component analysis (PCA), create new features as linear combinations of the original features. PCA is well-suited for regression problems such as this one because it reduces the dimensionality - and noise - of the data while still retaining the most important information. The ‘principal components’ always explain the greatest variance in the data, and, consequently, the first principal components often capture the data’s key patterns. The features’ coefficients for each component describes its contribution to that component. Accordingly, PCA eases interpretation of feature importance and pattern identification in data. Further, PCA is particular useful in machine learning workflows because it computationally efficient and it can help avoid overfitting data.

To determine the optimal number of principal components, a smaller number is preferable to avoid including components that do not substantially contribute to the variance in the data and instead contain a lot of noise. Therefore, the number of principal components is passed to the function as an argument (pca_comp) in the form of a list of possible number of principle components. The optimal number of components is then determined for each of the regressors using the GridSearchCV package, testing each number of components in the list and returning the number that gives the best performance on the training data.

## <u>Ensemble Learning<u>

Supervised learning is vital for this regression problem, as it predicts a continuous target variable using input features. In this case, various regression models are employed, utilising the principal components from PCA. Popular regression algorithms such as linear, ridge, lasso, decision trees, random forests, and gradient boosting machines can be considered, each with its strengths and weaknesses.

Cross-validated grid search is used to identify the best-performing model and its hyperparameters. Hyperparameters are adjustable settings that influence a model's learning process, such as the learning rate, regularisation strength, and tree depth. Keeping the hyperparameter space small is crucial, particularly for smaller datasets, as it reduces the risk of overfitting and ensures more efficient computational performance. The GridSearchCV package in scikit-learn evaluates each regressor's performance through cross-validation, providing a reliable estimate of the model's performance on unseen data.

The model with the best overall performance is selected and trained on the entire dataset, incorporating the optimal number of principal components. The final model can be used for making predictions and assessing feature importance, offering valuable insights into the relationships between input features and the target variable.By combining PCA with supervised regression algorithms, a more efficient machine learning workflow is achieved, helping to avoid overfitting, reduce computational complexity, and improve interpretability.

## <u>Core Function<u>

The text below details the core function used for modelling and evaluation in this project.

<u> Preprocessing & the Pipeline <u>

As noted during the data exploration, the significant number of null values in Accountability Percentile and PPI would hinder model fitting and necessitate the use of the KNN imputer in the preprocessing pipeline due to its ability to preserve patterns in the dataset. Furthermore, since PCA requires standardisation, a StandardScaler is employed. This scaler is more robust for features with different units of measurement or scales, giving them equal importance. As this dataset includes percentages as well as monetary expenditures as numeric variables, this is particularly beneficial. Additionally, given the wide ranges and outliers in the data, this standardiser effectively reduces the impact of outliers. As a result, the function incorporates these preprocessing elements in the pipeline, which, as detailed below, is used to loop over regressors and display results for model evaluation. Notably, the TransformedTargetRegressor is applied to transform the target variable, allowing for a more accurate representation of its relationship with the predictor variables.
    
<u> Looping over Regressors & Hyperparameter Tuning <u>

As mentioned earlier, the unsupervised and supervised learning methods are combined into pipelines. The second section of the function iterates over the main regressors. The TransformedTargetRegressor came as a suggestion from Chat GPT when working with code from practical 5 that returned an error. It is employed to adjust the target variable before fitting the regression models; the TransformedTargetRegressor is then passed to the GridSearchCV as the estimator for which hyperparameters need tuning. 10-fold cross validation is used to further ensure accuracy. The hyperparameter space is kept relatively small because there are relatively few data points, and a larger hyperparameter space would be more computationally expensive. GridSearchCV also utilises cross-validation with a specified number of folds, ensuring the model's performance is assessed on multiple data subsets to improve generalisation.
    
<u>Model Evaluation and Presentation<u>

After executing each regressor, the function generates several outputs. For each model, it presents the number of components utilised for PCA in model fitting after completing hyperparameter tuning, as well as the R2 and RMSE values indicating the model's performance on the test data. Moreover, for every model, the top 5 principal components are displayed along with the columns from the original dataframe that contribute to their variance, enhancing interpretability. An important aspect of the evaluation comes from the TransformedTargetRegressor because it is able to perform inverse transformation on the predictions. The inverse transformation refers to the process of converting the transformed target variable back to its original form. This facilitates more accurate and interpretable comparisons to be drawn and it ensures the R2 and RMSE are calculated using the original scale of the target variable. Lastly, a few comparative elements are provided. The first is a table comparing the R2 and RMSE values for the test and training data across all models. Following that, a bar chart illustrating the R2 and RMSE outputs from each model is shown for easy visual comparison.

```python
def models_with_pca(X_train, X_test, y_train, y_test, features_val, pca_comp):
    
    '''    
    NOTES:

    This function takes in training and testing data (predictor variables and target variable) as well as a list of potential numbers of 
    components for PCA. It starts by preprocessing the data: first, it imputes missing values in the target variable using KNNImputer, which 
    replaces missing values with the average of their k-nearest neighbours in the feature space. Next, it preprocesses predictor variables 
    using ColumnTransformer, applying the StandardScaler to numeric features.

    The function fits regression models to the preprocessed data using PCA, determining the ideal number of components through GridSearchCV. 
    It employs the TransformedTargetRegressor to modify the target variable without affecting predictor variables. Multiple regression models
    are run (Linear Regression, Random Forest Regressor, XGBoost Regressor, Decision Tree Regressor, and Gradient Boosting Regressor) with 
    PCA in the pipeline, and the ideal hyperparameters for each regressor are found using GridSearchCV. The performance of each regressor is
    evaluated on the testing data.

    Additionally, the function outputs visualisations of the R-squared and RMSE results, as well as the most important columns and their
    coefficients for each component in the PCA. It returns a list of fitted pipelines for each regressor after PCA and hyperparameter 
    optimisation. While the function can be applied to other datasets with similar types of questions, some adjustments to the hyperparameter
    space may be necessary due to the relatively small size of the current dataset.
    
    It is also important to note that this function is computationally expensive because it is comprehensive and streamlines several processes.
    
    Arguments:
        - X_train: A pandas DataFrame or numpy array, containing the training predictor variables
        - X_test: A pandas DataFrame or numpy array, containing the testing predictor variables
        - y_train: A pandas Series or numpy array, containing the training target variable
        - y_test: A pandas Series or numpy array, containing the testing target variable
        - features_val: A numpy array, the values in columns for the predictor variables in the regressions  
        - pca_comp: A list of integers, representing the number of potential components to use for PCA

    Returns:
        - A list of fitted pipelines for each regressor after PCA and hyperparameter optimisation.
        
    Credits:
        - Code for this function comes from Data Science for Spatial Systems Practicals 5 and 9
        - Troubleshooting when writing the code came from consulting Chat GPT on how to address errors that arose when writing the code.
          This was used predominantly in sorting out how to display the principle components so they could be better interpreted. Further,
          if consistantly encountering errors
    '''

# Preprocessing & the Pipeline

    # Add an imputer for the x and y values so that fitting the regressors can work and won't throw an error.
    # .ravel() flattens the data into a 1D array so it can be used to reshape the imputed y_train and y_test from 2D arrays into 1D arrays.
    y_imputer = KNNImputer()
    y_train = y_imputer.fit_transform(y_train.reshape(-1, 1)).ravel()
    y_test = y_imputer.transform(y_test.reshape(-1, 1)).ravel()
    
    # Change the features_val to integer indices
    features_indices = list(range(len(features_val)))

    # Setting-Up Transformer Pipeline
    numeric_transformer = Pipeline(steps=[('imputer', KNNImputer()),
                                          ('scaler', StandardScaler())])
    preprocessor = ColumnTransformer(
         transformers=[('num', numeric_transformer, features_indices)])

    regressors = {'Linear': LinearRegression(),
                  'RF': RandomForestRegressor(random_state=123),
                  'XGB': XGBRegressor(random_state=123),
                  'CART': DecisionTreeRegressor(random_state=123),
                  'GBR': GradientBoostingRegressor(random_state=123)}
    
    # Create an empty list to append the results of the function later
    model_results = [] 

# Looping over Regressors & Hyperparameter Tuning

    # The below code comes from Data Science for Spatial Systems Practical 5 (1.5) where several regressors are looped over to
    # compare their performance. However, this incorperates the hyperparameter tuning.
    
    for name, regressor in regressors.items():
        pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                   ('pca', PCA(random_state=123)),
                                   ('regressor', regressor)])

        transformed_regressor = TransformedTargetRegressor(regressor=pipeline)

        # Tune the hyperparameters and fit the models, using GridSearchCV with cross-validation of 5 folds
        # The hyperparmeter space is relatively small for this function because there are relatively few datapoints.
        # This can be changed for users who want to use this function with other data.
            
        if name == 'Linear':
            param_grid = {'regressor__pca__n_components': pca_comp}
        elif name == 'RF':
            param_grid = {'regressor__pca__n_components': pca_comp,
                          'regressor__regressor__n_estimators': [10, 50],
                          'regressor__regressor__max_depth': [2, 3, 5]}
        elif name == 'XGB':
            param_grid = {'regressor__pca__n_components': pca_comp,
                          'regressor__regressor__n_estimators': [10, 50],
                          'regressor__regressor__max_depth': [2, 3, 5]}
        elif name == 'CART':
            param_grid = {'regressor__pca__n_components': pca_comp,
                          'regressor__regressor__max_depth': [2, 3, 5]}
        elif name == 'GBR':
            param_grid = {'regressor__pca__n_components': pca_comp,
                          'regressor__regressor__n_estimators': [10, 50],
                          'regressor__regressor__max_depth': [2, 3, 5]}


        grid_search = GridSearchCV(transformed_regressor, param_grid=param_grid, cv=10, scoring='neg_mean_squared_error')
        grid_search.fit(X_train, y_train)
        y_train_pred = grid_search.predict(X_train)

# Model Evaluation and Presentation

        train_r2_score = r2_score(y_train, y_train_pred)
        train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))

        y_pred = grid_search.predict(X_test)
        best_n_components = grid_search.best_params_['regressor__pca__n_components']

        print(f"{name}:")
        print(f"Best number of components for {name}: {best_n_components}")
        print("Best Parameters: ", grid_search.best_params_)
        print(f"Testing Data R-squared score: {r2_score(y_test, y_pred)}")
        print(f"Testing Data RMSE: {np.sqrt(mean_squared_error(y_test, y_pred))}")
        print()

        # Since the goal of this project is both creating an accurate model *and* interpretability, it is essential to visualise the
        # principle components. Accoridngly, this code displays the principle components and the columns from the dataframe in them
        # in addition to the percent of the variance explained by them. 
        # Ideas seeded from: https://stackoverflow.com/questions/22984335/recovering-features-names-of-explained-variance-ratio-in-pca-with-sklearn
        
        # Get the fitted PCA object from the pipeline
        # Accessing steps of a scikit-learn pipeline:
        # https://stackoverflow.com/questions/28312534/how-to-access-steps-in-a-pipeline-after-fit
        pca = grid_search.best_estimator_.regressor_.named_steps['pca']

        # Rank the PCA components based on their importance. The absolute value of the PCA components represents the importance of 
        # each feature in each component. Higher absolute values indicate more significant contributions to the component.
        importance = np.abs(pca.components_)
        sorted_indices = np.argsort(importance, axis=1)[:, ::-1]

        # Limit the principal components to display in the final outcome of the function, not for tuning
        n_components_to_display = min(4, pca.n_components_)
        sorted_indices = sorted_indices[:, :n_components_to_display]

        # Get the most important column names and their coefficients as percentages for each component
        # Feature importance after PCA analysis:
        # https://stackoverflow.com/questions/50796024/feature-variable-importance-after-a-pca-analysis
        # Recovering features names of explained variance ratio in PCA with scikit-learn:
        # https://stackoverflow.com/questions/22984335/recovering-features-names-of-explained-variance-ratio-in-pca-with-sklearn
        most_important_columns = [
            [(features_val[sorted_indices[i, j]], importance[i, sorted_indices[i, j]] * 100)
                for j in range(n_components_to_display)]
            for i in range(n_components_to_display)]

        headers = [f"Component {i + 1}" for i in range(n_components_to_display)]
        ranked_columns = pd.DataFrame(most_important_columns, columns=headers)

        # Sort each column individually based on the percentages
        # Sorting a pandas DataFrame by a specific column:
        # https://stackoverflow.com/questions/37787698/how-to-sort-pandas-dataframe-from-one-column
        for col in ranked_columns.columns:
            ranked_columns[col] = ranked_columns[col].sort_values(ascending=False, key=lambda x: x.str[1]).tolist()

        ranked_columns['Component'] = range(1, n_components_to_display + 1)
        ranked_columns = ranked_columns.set_index('Component')

        print(f"Ranked important columns for {str(regressor).split('(')[0]} and Percentage of Component:")
        display(ranked_columns)
        print("---------------------------------------------")

        model_results.append({'Model': str(regressor).split('(')[0],
                      'Regressor': regressor,
                      'Best_n_components': best_n_components,
                      'R_squared_train': train_r2_score,
                      'R_squared_test': r2_score(y_test, y_pred),
                      'RMSE_train': train_rmse,
                      'RMSE_test': np.sqrt(mean_squared_error(y_test, y_pred))})

    # Create a DataFrame with the model results and display it
    results_df = pd.DataFrame(model_results)
    display(results_df)

    # Plotting the results
    fig, ax = plt.subplots(1, 2, figsize=(14, 5))

    models = results_df['Model']
    r2_train = results_df['R_squared_train']
    r2_test = results_df['R_squared_test']
    rmse_train = results_df['RMSE_train']
    rmse_test = results_df['RMSE_test']

    # Plot R-squared results
    ax[0].bar(models, r2_train, width=0.4, align='edge', label='Training')
    ax[0].bar(models, r2_test, width=-0.4, align='edge', label='Testing')
    ax[0].set_ylabel('R-squared')
    ax[0].set_title('Model Performance Comparison (R-squared)')
    ax[0].legend(loc='best')

    # Plot RMSE results
    ax[1].bar(models, rmse_train, width=0.4, align='edge', label='Training')
    ax[1].bar(models, rmse_test, width=-0.4, align='edge', label='Testing')
    ax[1].set_ylabel('RMSE')
    ax[1].set_title('Model Performance Comparison (RMSE)')
    ax[1].legend(loc='best')

    # Set x-axis labels rotation
    ax[0].set_xticklabels(models, rotation=45)
    ax[0].xaxis.set_major_locator(FixedLocator(range(len(models))))
    ax[1].set_xticklabels(models, rotation=45)
    ax[1].xaxis.set_major_locator(FixedLocator(range(len(models))))

    plt.show()
    
    # model_results is a list of dictionaries with the results of each model
    return model_results
```

```python
# Helper Function
def variance_bias_summary(model_results):
    '''
    This function takes the output of the 'models_with_pca' function and returns a pandas DataFrame containing
    the variance, bias, and total loss of the models.

    Arguments:
        - model_results: A list of dictionaries, each containing the results of a fitted model.

    Returns:
        - A pandas DataFrame with the variance, bias, and total loss of each model.
    '''
    summary = []

    for result in model_results:
        model = result['Model']
        variance = result['R_squared_train'] - result['R_squared_test']
        train_error = result['RMSE_train']
        test_error = result['RMSE_test']
        bias = train_error - test_error # This is because bias is the difference between the avergae prediction of the model and the correct value
        total_loss = variance + bias
        
        summary.append({'Model': model, 'Variance': variance, 'Train Error': train_error, 'Test Error': test_error, 'Bias': bias, 'Total Loss': total_loss})

    summary_df = pd.DataFrame(summary)
    
    return summary_df
```

```python
def bias_variance_analysis(model_results, X_train, X_test, y_train, y_test, num_rounds=50, random_seed=123):
    columns = ['Expected Loss - Train', 'Expected Loss - Test', 'Diff Loss', 'Bias - Train', 'Variance - Train', 'Bias - Test', 'Variance - Test']
    results = []

    for model_result in model_results:
        name = model_result['Model']
        regressor = model_result['Regressor']
        
        train_metrics = bias_variance_decomp(regressor, X_train, y_train, X_train, y_train, loss='mse', random_seed=random_seed, num_rounds=num_rounds)
        test_metrics = bias_variance_decomp(regressor, X_train, y_train, X_test, y_test, loss='mse', random_seed=random_seed, num_rounds=num_rounds)
        
        diff_loss = train_metrics[0] - test_metrics[0]
        model_data = [name] + list(train_metrics) + list(test_metrics) + [diff_loss]
        results.append(model_data)

    df_results = pd.DataFrame(results, columns=['Model'] + columns)
    df_results.set_index('Model', inplace=True)

    return df_results
```

```python
# calculating VIF
# This function comes from Data Science for Spatial Systems practical 6 which is amended from https://stackoverflow.com/a/51329496/4667568

def drop_column_using_vif_(df, list_var_not_to_remove=None, thresh=5):
    '''
    Calculates VIF each feature in a pandas dataframe, and repeatedly drop the columns with the highest VIF
    A constant must be added to variance_inflation_factor or the results will be incorrect

    :param df: the pandas dataframe containing only the predictor features, not the response variable
    :param list_var_not_to_remove: the list of variables that should not be removed even though it has a high VIF. For example, dummy (or indicator) variables represent a categorical variable with three or more categories.
    :param thresh: the max VIF value before the feature is removed from the dataframe
    :return: dataframe with multicollinear features removed
    '''
    while True:
        # adding a constatnt item to the data
        df_with_const = add_constant(df)

        vif_df = pd.Series([variance_inflation_factor(df_with_const.values, i) 
               for i in range(df_with_const.shape[1])], name= "VIF",
              index=df_with_const.columns).to_frame()

        # drop the const as const should not be removed
        vif_df = vif_df.drop('const')
        
        # drop the variables that should not be removed
        if list_var_not_to_remove is not None:
            vif_df = vif_df.drop(list_var_not_to_remove)
            
        print('Max VIF:', vif_df.VIF.max())
        
        # if the largest VIF is above the thresh, remove a variable with the largest VIF
        if vif_df.VIF.max() > thresh:
            # If there are multiple variables with the maximum VIF, choose the first one
            index_to_drop = vif_df.index[vif_df.VIF == vif_df.VIF.max()].tolist()[0]
            print('Dropping: {}'.format(index_to_drop))
            df = df.drop(columns = index_to_drop)
        else:
            # No VIF is above threshold. Exit the loop
            break

    return df
```

```python
# PREPROCESSING AND VIF 

# Drop the 'Progress and Performance Index (PPI) - All Students' column
df2 = df1.drop('School Accountability Percentile (1-99)', axis=1)

# Separate the independent and dependent variables
predict_cols = df2.columns[:-1]  # assuming the last column is the target column
target_col = df2.columns[-1]

X = df2[predict_cols]
y = df2[target_col]

# Impute missing values in the target variable
y_imputer = KNNImputer(n_neighbors=5)
y = y_imputer.fit_transform(y.to_numpy().reshape(-1, 1)).ravel()

# Preprocess the data using the pipeline
preprocessing = Pipeline([
    ('imputer', KNNImputer(n_neighbors=5)),
    ('scaler', StandardScaler())
])

X_preprocessed = preprocessing.fit_transform(X)

# Remove variables with high VIF
X_vif_removed = drop_column_using_vif_(pd.DataFrame(X_preprocessed, columns=predict_cols))

# Fit the linear regression model
lr_model = LinearRegression()
lr_model.fit(X_vif_removed, y)

print('---------------------------')

# Calculate and print the mean squared error and R^2 score
y_pred = lr_model.predict(X_vif_removed)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print('The results of Linear Regression Following VIF')
print(f'Mean squared error: {mse}')
print(f'R^2 score: {r2}')
```

```python
# CREATE A REGRESSION TABLE TO DISPLAY THE RESULTS

# Create a regression table showing the coefficients, standard errors, t-values, and p-values
X_vif_removed_sm = sm.add_constant(X_vif_removed)
lr_model_sm = sm.OLS(y, X_vif_removed_sm).fit()

#extract the key details from the regression to be use in the regression table
coef_names = ['const'] + list(X_vif_removed.columns)
coef_vals = [lr_model_sm.params[0]] + lr_model_sm.params[1:].tolist()
stderr_vals = [lr_model_sm.bse[0]] + lr_model_sm.bse[1:].tolist()
t_vals = [lr_model_sm.tvalues[0]] + lr_model_sm.tvalues[1:].tolist()
p_vals = [lr_model_sm.pvalues[0]] + lr_model_sm.pvalues[1:].tolist()

# Create the regression table and add the stars column
reg_table = pd.DataFrame({
    'Variable': coef_names,
    'Coefficient': coef_vals,
    'Std. Error': stderr_vals,
    't-value': t_vals,
    'p-value': p_vals})

# Filter the table to show only statistically significant variables
significant_vars = reg_table[reg_table['p-value'] < 0.05]

# Sort the significant variables in ascending order
significant_vars = significant_vars.sort_values('p-value', ascending=True)

print('Variables of Regression post-VIF with p-values under 0.05')
significant_vars
```
